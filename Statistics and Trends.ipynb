```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as pltim
import seaborn as sns
import numpy as np 

```

```python
# Load datasets
match_results = pd.read_csv('match_results_for_icc_mens_t20_world_cup_2024.csv')
batting_data = pd.read_csv('batting_stats_for_icc_mens_t20_world_cup_2024.csv')
bowling_data = pd.read_csv('bowling_stats_for_icc_mens_t20_world_cup_2024.csv')
```

```python
win_counts = match_results['Winner'].value_counts()
print("Number of Wins by Each Team:")
print(win_counts)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', 
          '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'] 

plt.figure(figsize=(10, 6))
win_counts.plot(kind='barh', color=colors[:len(win_counts)])
plt.title("Total Wins by Team")
plt.xlabel("Number of Wins")
plt.ylabel("Team")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(batting_data['Runs'], batting_data['SR'], color='dodgerblue', alpha=0.7, edgecolors='k')
plt.title("Runs vs. Strike Rate (Batting Stats)")
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
bowling_metrics = bowling_data[['Balls', 'Runs', 'Wkts', 'Ave', 'Econ', 'SR']]
bowling_metrics = bowling_metrics.replace('-', np.nan)
bowling_metrics = bowling_metrics.apply(pd.to_numeric, errors='coerce')
bowling_metrics = bowling_metrics.dropna()
correlation_matrix = bowling_metrics.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap of Bowling Metrics")
plt.show()
```
